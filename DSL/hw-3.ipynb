```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.datasets import load_boston

import warnings
warnings.filterwarnings('ignore')
```

```python
boston = load_boston()
X = pd.DataFrame(boston['data'], columns=boston['feature_names'])
y = pd.DataFrame(boston['target'], columns=['price'])

X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
df = pd.DataFrame({'y_test': y_test['price'], 'y_pred': y_pred.flatten()})
df.head()
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred)
```

```python
X = pd.DataFrame(df, columns=df.columns[:-1])
X.head()
```

```python
y = pd.Series(df['Class'].squeeze())
type(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=100, test_size=0.3, stratify=y)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
parameters = [{'n_estimators': [10, 15],
               'max_features': np.arange(3, 5),
               'max_depth': np.arange(4, 7)}]

clf = GridSearchCV(
    estimator=RandomForestClassifier(random_state=100),
    param_grid=parameters,
    scoring='roc_auc',
    cv=3
)
```

```python
clf.fit(X_train, y_train)
```

```python
clf.best_params_
```

```python
y_pred_proba = clf.predict_proba(X_test)[:, 1]
y_pred_proba
```

```python
from sklearn.metrics import roc_auc_score
```

```python
roc_auc_score(y_test, y_pred_proba)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#8
plt.figure(figsize=(15, 10))

sns.set(font_scale=1.4)

X_corr = X.corr()
X_corr = np.round(X_corr, 2)
X_corr[np.abs(X_corr) < 0.5] = 0

sns.heatmap(X_corr, annot=True, linewidths=.5, cmap='coolwarm')

plt.title('Correlation matrix')
plt.show()
```

```python
#9

x_corr_indexes = X_corr['target'].index
print(x_corr_indexes)
high_corr = []
for i in x_corr_indexes[:-1]:
    if X_corr['target'][i] != 0:
        high_corr.append(i)
high_corr
```

```python
#10
X.drop('target', axis=1, inplace=True)
X.head()
```

```python
for i in high_corr:
    X[i+'_2'] = None
    X[i+'_2'].fillna(X[i] ** 2, inplace=True)
X.head()
```
